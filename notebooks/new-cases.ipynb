```python
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import altair as alt
```

```python
url = "http://lahub.maps.arcgis.com/sharing/rest/content/items/8aba663239fc428f8bcdc48e213e3172/data"
df = pd.read_csv(url)
```

```python
df = df[['county', 'state', 'fips', 'date', 'Lat', 'Lon', 'cases',
       'deaths', 'state_cases', 'state_deaths', 'new_cases', 'new_deaths',
       'new_state_cases', 'new_state_deaths']]
```

```python
# Plot 3, 5, 7-day moving average
df['date'] = pd.to_datetime(df.date)

df['avg3'] = df.new_cases.rolling(window=3).mean()
df['avg5'] = df.new_cases.rolling(window=5).mean()
df['avg7'] = df.new_cases.rolling(window=7).mean()
```

```python
alt.Chart(df[(df.county=="Los Angeles") & (df.date >= '3/1/20')]).mark_line().encode(
    x='date',
    y='new_cases'
)
```

```python
alt.Chart(df[(df.county=="Los Angeles") & (df.date >= '3/1/20')]).mark_line().encode(
    x='date',
    y='avg3'
)
```
