```python
import geopandas as gpd
```

```python
data_locations = {"2017-Q1": "https://opendata.arcgis.com/datasets/699eedf6140943e6b7fbea068a160a4c_0.geojson", 
                  "2017-Q4": "https://opendata.arcgis.com/datasets/9fe1b43232194e43acfdde66fc90aca0_0.geojson"}
```

```python
data = {}
for key, value in data_locations.items():
    data[key] = gpd.read_file(value)
    
```

```python
data["2017-Q4"] = data["2017-Q4"].rename({"BIScore":"BulkyScore"}, axis=1)
```

```python
import pandas as pd
pd.concat(data)
```

```python
data["2017-Q4"].plot()
```
